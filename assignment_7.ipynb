## Question 3
Univariate Analysis

```python
sns.boxplot(y='Age',data=df)
```

```python
sns.countplot(x='Pclass',data=df ,hue='Sex', palette='Reds_d')
```

```python
plt.hist(df['Age'])
```

```python
sns.distplot(df['Age'],kde=True, color="y")
```

```python
sns.distplot(df['Age'],kde=True, color="black", hist=False, rug=True)
```

```python
sns.barplot(y='Pclass',data=df)
```

```python
sns.kdeplot(df['Pclass'], color="b", shade=True)
```

```python
var = df[(df.Sex.str.match('male'))&(df.Age>62)]        #
var
```

## We found by the help of boxplot that males above the age of 62 have no chance of surviving so we are eliminating them from the dataset.
## Removing Outliers

```python
df.drop(var.index)
```

## Question 4
Bivariate Analysis

```python
plt.plot(df["Age"],df["Fare"])
plt.ylabel("Age")
plt.xlabel("Fare")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.kdeplot(df["Age"],df["Fare"])
```

```python
sns.jointplot(df["Age"],df["Fare"],kind="hex",color="g")
```

```python
sns.barplot(df["Age"],df["Fare"],hue="Sex",data=df,color="red")
```

```python
sns.lmplot("Fare","Age",df,hue="Sex",fit_reg=False)
```

```python
sns.stripplot(x="Pclass",y="Fare",data=df)
```

```python
sns.swarmplot(df["Pclass"],df["Fare"],data=df)
```
